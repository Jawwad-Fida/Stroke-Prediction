```python
# Visualization libraries
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

# Computational libraries
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("healthcare-dataset-stroke-data.csv")
```

```python
data.head()
```

## Pie Chart - Proportion of Stroke Samples (Label Class)

```python
import plotly.express as ex

fig = ex.pie(data,names='stroke')
fig.update_layout(title='<b>Proportion Of Stroke Samples<b>')
fig.show()
```

Here, we can see that our dataset is imbalanced. We likely have to use sampling techniques to make the data balance.

```python
sns.set_theme(style="darkgrid")
```

## Finding correlation to class variable using Heatmap

```python
plt.figure(figsize=(16,8))
sns.heatmap(data.corr(),cmap="Greens");
```

## Count Plot - Gender

```python
sns.set(rc={'figure.figsize':(8,6)})
ax = sns.countplot(data=data, x="gender")
plt.show()
```

Above, you can see the Females present in our dataset is higher than males.

## Count Plot - Worktype

```python
sns.set(rc={'figure.figsize':(8,6)})
ax = sns.countplot(data=data, x="work_type")
plt.show()
```

A lot of people works in Private sector.

## Pie Chart - Smoking Status

```python
fig = ex.pie(data,names='smoking_status')
fig.update_layout(title='<b>Proportion Of Different Smoking Categories Among Stroke Population<b>')
fig.show()
```

A lot of people never smoked in their life. But, we also don't know the exact status of Unknowns in our dataset.

## No Stroke vs Stroke by Age


```python
plt.figure(figsize=(12, 10))

sns.distplot(data[data['stroke'] == 0]["age"], color='green') # No Stroke - green
sns.distplot(data[data['stroke'] == 1]["age"], color='red') # Stroke - Red

plt.title('No Stroke vs Stroke by Age', fontsize=15)
plt.xlim([18,100])
plt.show()
```

This confirms what our intuitions told us. The older you get, the more at risk you get.

However, you may have noticed the low risk values on the y-axis. This is because the dataset is highly imbalanced.

## No Stroke vs Stroke by BMI

```python
plt.figure(figsize=(12,10))

sns.distplot(data[data['stroke'] == 0]["bmi"], color='green') # No Stroke - green
sns.distplot(data[data['stroke'] == 1]["bmi"], color='red') # Stroke - Red

plt.title('No Stroke vs Stroke by BMI', fontsize=15)
plt.xlim([10,100])
plt.show()
```

From the graph, it shows that the density of overweight people who suffered a stroke is more.

## No Stroke vs Stroke by Avg. Glucose Label

```python
plt.figure(figsize=(12,10))

sns.distplot(data[data['stroke'] == 0]["avg_glucose_level"], color='green') # No Stroke - green
sns.distplot(data[data['stroke'] == 1]["avg_glucose_level"], color='red') # Stroke - Red

plt.title('No Stroke vs Stroke by Avg. Glucose Level', fontsize=15)
plt.xlim([30,330])
plt.show()
```

From graph, it shows that the density of people having glucose level less than 100 suffered stroke more.
